```python
import matplotlib.pyplot as plt
import numpy as np
import MDAnalysis as mda 
import MDAnalysis.auxiliary 
```

```python
potential_file = "../minimisation/potential.xvg"
reader = mda.auxiliary.XVG.XVGReader(potential_file)
time = [step.data[0] for step in reader]
energy = [step.data[1] for step in reader]
plt.plot(time, energy, c="red")
plt.xlabel("time / ps")
plt.ylabel(r"Energy / kJ mol$^{-1}$")
plt.tight_layout()
plt.savefig("../plots/potential.png")
```

```python
pulling_file = "../pulling_nvt/pulling_pullx.xvg"
reader = mda.auxiliary.XVG.XVGReader(pulling_file)
time = [step.data[0] for step in reader]
x1 = [step.data[1] for step in reader]
x2 = [step.data[2] for step in reader]
plt.plot(time, x1, label="x1")
plt.plot(time, x2, label="x2")
plt.xlabel("time / ps")
plt.ylabel("Distance / nm")
plt.legend()
plt.tight_layout()
plt.savefig("../plots/nvt_pulling.png")
```
