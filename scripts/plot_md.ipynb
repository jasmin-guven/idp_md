```python
import matplotlib.pyplot as plt
import numpy as np
import MDAnalysis as mda 
import MDAnalysis.auxiliary 
```

```python
potential_file = "../minimisation/potential.xvg"
reader = mda.auxiliary.XVG.XVGReader(potential_file)
print(reader)
```

```python
time = []
energy = []
for step in reader:
    time.append(step.data[0])
    energy.append(step.data[1])  
```

```python
plt.plot(time, energy, c="red")
plt.xlabel("time / ps")
plt.ylabel(r"Energy / kJ mol$^{-1}$")
plt.tight_layout()
plt.savefig("../plots/potential.png")
```

```python
temperature_file = "../nvt/nvt.xvg"
reader = mda.auxiliary.XVG.XVGReader(temperature_file)
print(reader)
```
